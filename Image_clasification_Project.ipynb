```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# For Jupyter Notebook inline plotting
%matplotlib inline
```

```python
# using pandas to read the database store in the same folder
import pandas as pd
data = pd.read_csv("mnist.csv")
# viewing coloumn heads
data.head()
```

```python
In [5]: #extracting data from the dataset and viewing them up close 
a = data.iloc[3,1:].values
```

```python
# reshaping the extracted data into a reasonable size
a = a.reshape(28,28).astype('uint8')
plt.imshow(a)
plt.imshow(a, cmap='gray')
plt.title("Digit Image")
plt.axis('off')  # Optional: removes axis ticks
plt.show()
```

```python
#preparing the data
#separating the data
df_x = data.iloc[:,1:]
df_y = data.iloc[:,0]
```

```python
#creating test and train sizes/batches
x_train,x_test,y_train,y_test = train_test_split(df_x,df_y,test_size = 0.2,random_state = 4)
```

```python
#cheack data
y_train.head()
```

```python
# call rf classifier
rf = RandomForestClassifier(n_estimators = 100)
```

```python
# fit the model
rf.fit(x_train,y_train)
```

```python
# predication on test data
pred = rf.predict(x_test)
```

```python
pred 
```

```python
# check prediction accuracy
s = y_test.values
# calculate number of corectly predicted values
count = 0
for i in range (len(pred)):
    if pred[i] == s[i]:
        count = count+1
```
